# Inplace Operations

```python
m = torch.rand(2,3)
n = torch.rand(2,3)
print(m)
print(n)
```

```python
m.add_(n)
```

```python
m
```

```python
n
```

```python
torch.relu(m) # this would not have dont changes to the tensor itself. it will be a copy
```

```python
# in order to do permanent changes by just adding a '_' underscore infront of the paranthese while calling a function
torch.relu_(m)
```

```python
m
```
