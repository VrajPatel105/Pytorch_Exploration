# Autograd

Autograd is a core component of PyTorch taht provides automatic differentiaion for tensor operations.

It enables gradient computation, which is essential for training machine learning models using optimization algorithms like gradient descent

```python
import torch
```

```python
x = torch.tensor(3.0, requires_grad=True) # scaler tensor for now
```

```python
y = x**2
```

```python
y
```

```python
y.backward()
```

```python
x.grad
```

```python
# another example
x = torch.tensor(3.0, requires_grad=True)
```

```python
y = x ** 2
```

```python
z = torch.sin(y)
```

```python
print(x)
print(y)
print(z)
```

```python
z.backward()
```

```python
x.grad
```

# Clearing Gradients

When we call gradients multiple times, the gradients starts accumulating, which is not a good thing. We have to clear it


` We clear gradients every epochs`

```python
x.grad
```

```python
x.grad.zero_()
```
